# Importing librairies used for the sentiment analysis

```python
## Import entire libraries

import pandas as pd
import numpy as np

#library for plotting stuff
import matplotlib.pyplot as plt
#so that plots appear in the same browser window
%matplotlib inline


## Import specific items only from the sklearn library

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
```

# Reading Data From two files

```python
# Read sentiment data from two different files
data_pos = []
data_neg = []

with open('./pos_comm.txt', 'r', encoding='utf-8') as fid:
    for line in fid:
        data_pos.append(line)

with open('./bad_comm.txt', 'r', encoding='utf-8') as fid:
    for line in fid:
        data_neg.append(line)
```

```python
print(type(data_pos))
print([len(data_pos), len(data_neg)])
```

# Creating X and Y variables for the Data

```python
# X contains all documents
X = data_pos + data_neg
print(type(X))
```

```python
# y contains the sentiment category of all documents
y = [1]*len(data_pos) + [0]*len(data_neg)
print(type(y))
```

```python
print(len(X))
print(len(y))
```

# Test and Training Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=321)
```

```python
print([len(X_train), len(y_train)])
print([len(X_test), len(y_test)])
print(len(X_test)/len(X))
```

# Sentiment Analysis

```python
count_vect = CountVectorizer()

# the fit() method tokenizes documents and extracts their vocabulary, i.e. set of distinct words
count_vect.fit(X_train)
len(count_vect.get_feature_names())
```

```python
# calculate the document-term matrix for the training corpus
X_train_vectorized = count_vect.transform(X_train)
print(type(X_train_vectorized))
print(X_train_vectorized.shape)
```

```python
# determine number of non-zero elements in document-term matrix of training corpus
X_train_vectorized.nnz
```

# Analysis of the Distribution of document frequency

```python
# Calculate document frequency of each vocabulary word
doc_freq = np.array((X_train_vectorized > 0).sum(0)).ravel()

print(doc_freq.shape)
print([np.amin(doc_freq), np.median(doc_freq), np.mean(doc_freq), np.amax(doc_freq)])
```

```python
# Plot histogram of document frequency
plt.hist(doc_freq)
```

```python
# Calculate relative document frequency of each vocabulary word
n,m = X_train_vectorized.shape
rel_doc_freq = np.array((X_train_vectorized > 0).sum(0)/n).ravel()

print(rel_doc_freq.shape)
print([np.amin(rel_doc_freq), np.median(rel_doc_freq), np.mean(rel_doc_freq), np.amax(rel_doc_freq)])
```

```python
# Which words have high document frequency?
vocab = count_vect.get_feature_names()
[x for index,x in enumerate(vocab) if doc_freq[index]>200]   # THIS Python structure IS CALLED A COMPREHENSION
```

# Remove documents of length 0

```python
# Which documents have length 0? (i.e. contain 0 words)
idx = np.where(words_per_doc == 0)[0]
[X_train[i] for i in idx]
```

```python
## Select training documents that have length 0
idx2 = np.where(words_per_doc > 0)[0]
X_train2 = [X_train[i] for i in idx2]
y_train2 = [y_train[i] for i in idx2]
print([len(X_train), len(X_train2)])
print([len(y_train), len(y_train2)])
```

```python
# overwrite the training corpus
X_train = X_train2
y_train = y_train2
```

```python
# re-compute the document-term matrix
X_train_vectorized = count_vect.transform(X_train)
X_train_vectorized.shape
```

```python
# verify results: re-compute number of words in each document
words_per_doc = np.array((X_train_vectorized > 0).sum(1))
print(words_per_doc.shape)
print(np.amin(words_per_doc))
print(np.amax(words_per_doc))
```

# Classification model

```python
# Train the model using LR method
LR_model = LogisticRegression()
LR_model.fit(X_train_vectorized, y_train)
```

```python
# first obtain the document-term matrix of the test corpus
X_test_vectorized = count_vect.transform(X_test)
print(type(X_test_vectorized))
print(X_test_vectorized.shape)
```

```python
# Use this model to predict the sentiment category of test documents
LR_predictions = LR_model.predict(X_test_vectorized)
print(type(LR_predictions))
print(LR_predictions.shape)
```

```python
print(LR_predictions[:30])
print(np.array(y_test[:30]))
np.sum(LR_predictions[:1000] == np.array(y_test[:1000]))
```

```python
# Calculate classification rate on the test corpus

LR_classif_rate = accuracy_score(y_test, LR_predictions)
LR_classif_rate
```

# Interpretation

```python
# get the LR model's coefficients (there is one coefficient per vocabulary word)
LR_coefs = LR_model.coef_[0]
print(LR_coefs.shape)

# Sort coefficient values in ascending order
idx = LR_coefs.argsort()  #sort by actual value
idx2 = abs(LR_coefs).argsort()  #sort by absolute value

# Display the 20 smallest and 20 largest coefficients

feature_names = np.array(count_vect.get_feature_names())

print('20 Most negative coefficients (significant for negative sentiment):\n')
print(' Coefficient values:\n{}\n'.format(LR_coefs[idx[:20]]))
print(' Feature names:\n{}\n'.format(feature_names[idx[:20]]))

print("--------------------------------------------------------------")

print('20 Most positive coefficients (significant for positive sentiment):\n')
print(' Coefficient values: \n{}'.format(LR_coefs[idx[:-21:-1]]))
print(' Feature names: \n{}'.format(feature_names[idx[:-21:-1]]))

print("--------------------------------------------------------------")

print('20 Coefficients closest to zero (irrelevant features) :\n')
print(' Coefficient values:\n{}\n'.format(LR_coefs[idx2[:20]]))
print(' Feature names:\n{}\n'.format(feature_names[idx2[:20]]))
```
